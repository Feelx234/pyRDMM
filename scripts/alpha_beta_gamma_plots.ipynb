```python
import seaborn as sns
import pandas as pd
```

```python
def set_colums(df, ex_names = ['Like','Log'], sim_names = ['Like','Log','Cooks','Par']):
    cols = []
    for ex in ex_names:
        for sim in sim_names:
            for t in ['mean','std']:
                cols.append((ex,sim,t))
    df.columns = pd.MultiIndex.from_tuples(cols,names=['ex','sim','type'])
```

```python


df = pd.read_csv('E:/Masterarbeit/regression_exhaustive_results_top10.csv',delim_whitespace=True,header=None)
df.set_index([0,1,2],inplace=True)
set_colums(df)
df_1 =pd.read_csv('E:/Masterarbeit/regression_exhaustive_results_top10_1.csv',delim_whitespace=True,header=None)
df_1.set_index([0,1,2],inplace=True)
set_colums(df_1,['Cooks','Par'])
```

```python
def drop_like_log(df):
    df.drop([('Like','Cooks','mean'),('Like','Cooks','std')],axis=1,inplace=True)
    df.drop([('Like','Log','mean'),('Like','Log','std')],axis=1,inplace=True)
    df.drop([('Log','Log','mean'),('Log','Log','std')],axis=1,inplace=True)
    df.drop([('Log','Cooks','mean'),('Log','Cooks','std')],axis=1,inplace=True)
    df.drop([('Log','Like','mean'),('Log','Like','std')],axis=1,inplace=True)
    df.drop([('Log','Par','mean'),('Log','Par','std')],axis=1,inplace=True)
drop_like_log(df)
len(df.columns)
```

```python
def drop_cooks_var(df_1):
    df_1.drop([('Par','Cooks','mean'),('Par','Cooks','std')],axis=1,inplace=True)
    df_1.drop([('Par','Log','mean'),('Par','Log','std')],axis=1,inplace=True)
    df_1.drop([('Cooks','Log','mean'),('Cooks','Log','std')],axis=1,inplace=True)
    df_1.drop([('Cooks','Cooks','mean'),('Cooks','Cooks','std')],axis=1,inplace=True)
    df_1.drop([('Cooks','Like','mean'),('Cooks','Like','std')],axis=1,inplace=True)
    df_1.drop([('Cooks','Par','mean'),('Cooks','Par','std')],axis=1,inplace=True)
drop_cooks_var(df_1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_df(df, path):
    x_names=list(map(str,df.index.to_list()))
    i=0
    lasty = None
    plt.figure(figsize=(12,9))
    x=np.arange(len(x_names))
    for name,content in df.iteritems():
        i+=1
        if i % 2 == 1:
            lasty=content
        if i % 2 == 0:
            plt.errorbar(x=x,y=lasty/100, yerr=content/100,capsize=4,fmt='x',label="ex="+name[0]+", sim="+name[1])

    plt.axvline(3.5,color='k',linestyle=':')
    plt.axvline(7.5,color='k',linestyle=':')
    plt.axvline(11.5,color='k',linestyle='--',linewidth=2)
    plt.axvline(15.5,color='k',linestyle=':')
    plt.axvline(19.5,color='k',linestyle=':')
    plt.axvline(23.5,color='k',linestyle='--',linewidth=2)
    plt.axvline(27.5,color='k',linestyle=':')
    plt.axvline(31.5,color='k',linestyle=':')
    plt.xticks(x, x_names, rotation='vertical');
    plt.ylabel('accuracy')
    plt.xlabel(r'$(\gamma, \beta, \alpha)$')
    plt.ylim([-0.1,1.1])
    plt.legend(loc=4)
    plt.savefig(path,dpi=300)
```

```python
plot_df(df,'E:/Masterarbeit/Like.png')
```

```python
plot_df(df_1, 'E:/Masterarbeit/Var.png')
```
