```python
coss=np.abs(np.cos(phi))
tans=np.tan(phi)
for i in range(100):
    b=tans[i]
    c=coss[i]
    x=np.linspace(-c,c)
    plt.plot(x,b*x)
plt.xlabel('x')
plt.ylabel('x * b')
plt.savefig(r'E:\Masterarbeit\thesis-template\images\from-phi.pdf')
```

```python
bmax=np.abs(np.tan(5 / 360 * np.pi+np.pi/2))
bs=2*bmax*np.random.rand(100)-bmax
cs=np.cos(np.arctan(bs))
for i in range(100):
    b=bs[i]
    c=cs[i]
    x=np.linspace(-c,c)
    plt.plot(x,b*x)
plt.xlabel('x')
plt.ylabel('x * b')
```
